```
# Read the original image in BGR format: 

# Parameter for imread
# 1: Loads a color image
# 0: Loads image in grayscale mode
image = cv2.imread(path+'natural_living.bmp',1) 

# Show the BGR image:
plt.title('RGB image in BGR format')
plt.imshow(image)
plt.show()

# Convert from the BGR to RGB
RGBimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show the RGB image:
plt.title('RGB image in RGB format')
plt.imshow(RGBimage)
# plt.show()
```

```
# Read the original image in BGR format: 
image = cv2.imread(path+'natural_living.bmp',0) 

# Show the grayscale image:
plt.title('Grayscale image in BGR format')
plt.imshow(image)
plt.show()

# Convert from the BGR to RGB
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show the grayscale image:
plt.title('Grayscale image in RGB format')
plt.imshow(grayimage)
plt.show()
```

```
# Create a grayscale image with NumPy

# Parameter for numPy
# (height,width) for creating a grayscale image
# (height,width,channel) for a color image

# Create an image
image = np.ones([8,8], np.uint8)

# Edit pixel values in image from 1 to 255
image = image*255

print(image)
plt.imshow(image,cmap = 'gray', vmin=0, vmax=255)

# Access and edit pixel values by row and column coordinates
image[:,4:6] = 0
image[1,3] = 0
image[2,2:4] = 0

# Show the grayscale image:
plt.title('grayscale image')
plt.imshow(image,cmap = 'gray', vmin=0, vmax=255)
plt.show()
```

```
# Create a grayscale image with NumPy

# Parameter for numPy
# (height,width) for creating a grayscale image
# (height,width,channel) for a color image

# Create an image
image = np.ones([8,8,3], np.uint8)

# Edit pixel values in image from 1 to 255
image = image*255

# print(image)
# plt.imshow(image,cmap = 'gray', vmin=0, vmax=255)

# Access and edit pixel values by row and column coordinates

# Red
image[:,4:6,0] = 255
image[1,3,0] = 255
image[2,2:4,0] = 255

# Green
image[:,4:6,1] = 0
image[1,3,1] = 0
image[2,2:4,1] = 0

# Blue
image[:,4:6,2] = 0
image[1,3,2] = 0
image[2,2:4,2] = 0

# Show the grayscale image:
plt.title('grayscale image')
plt.imshow(image)
plt.show()
```
